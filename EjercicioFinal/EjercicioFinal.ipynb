```python
import pandas as pd
```

```python
#Conseguir csv de la web y guardarlo en un DataFrame
df_ti = pd.read_csv("https://raw.githubusercontent.com/omarmendoza564/datos/main/datos/titanic.csv")
df_ti
```

```python
#Guardar el contenido del DataFrame en un archivo csv local 
df_ti.to_csv("titanic.csv")
```

#### Limpieza de datos

##### 1. Si la edad es null (na) llenar el campo con la media de la edad (usar el metodo fillna())

```python
df_ti['age'].describe()
```

```python
df_ti['age']= df_ti.age.fillna(df_ti.age.mean())
df_ti
```


##### 2. Si embarqued es na llenar con S


```python
df_ti['embarked']= df_ti.embarked.fillna('S')
df_ti
```

##### 3. Cambiar los valores alfanumericos de sex a numero usando { 'female': 1, 'male': 0 }


```python
df_ti['sex'] = df_ti['sex'].replace({'female' : 1, 'male' : 0})
df_ti
```

##### 4. Cambiar los valores alfanumericos de embarked a numero usando { 'S': 0, 'C': 1, 'Q': 2 }


```python
df_ti['embarked'] = df_ti['embarked'].replace({'S':0, 'C':1, 'Q':2})
df_ti
```

##### 5. Crear una nueva variable llamada child con valor 1 para los pasajeros con una edad menor a 16 años y 0 si el pasajero tiene 16 años o mas


```python
df_ti['child'] = (df_ti['age'] < 16).astype(int)
df_ti
```

##### 6. Eliminar las columnas ticket, cabin y name


```python
df_ti = df_ti.drop(['ticket', 'cabin', 'name'], axis=1)
df_ti
```

##### 7. Validar que no existan registros duplicados


```python
dup=df_ti.duplicated(subset=['sex','age','sibsp', 'parch','fare','embarked'],keep='last')
dup
```

```python
df_ti[dup]
```

```python
#Hay 232 registros duplicados, para tirarlos corremos el la siguiente línea
df_ti.drop_duplicates(subset=['sex','age','sibsp', 'parch','fare','embarked'],keep='last', inplace=True)
df_ti.count()
```

##### 8. Listar los valores unicos del campo sex


```python
df_ti['sex'].unique()
```

```python
df_ti['sex'].value_counts(dropna=False)
```

##### 9. Quitar los espacios vacios de los campos que no sean numericos


```python
df_ti.dtypes
```

```python
#boat, home.dest
df_ti['boat'] = df_ti['boat'].str.strip()
df_ti['home.dest'] = df_ti['home.dest'].str.strip()
df_ti
```
